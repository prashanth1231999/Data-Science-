Q. 2) In which year the highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
df
```

```python
df["Release_Date"]=pd.to_datetime(df["Release_Date"])
```

```python
df.groupby(df.Release_Date.dt.year,as_index=False)[["Release_Date","Show_Id"]].count()
```

```python
l=[]
for i,j in df.groupby(df.Release_Date.dt.year)["Show_Id"]:
    d={}
    d["Year"]=i
    d["no_of_shows"]=j.count()
    l.append(d)
a=pd.DataFrame(l,index=[i for i in range(14)])
a
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data=a,x="Year",y="no_of_shows")
```

Q. 11) Which individual country has the Highest No. of TV Shows ?

```python
df.isna().sum()
a=df[["Country"]].fillna("")
```

```python
a.isna().sum()
```

```python
l=list(a.Country.unique())
l
```

```python
l1=[i.split(",") for i in l]
```

```python
l2=[]
for i in l1:
    for j in i:
        if j!="":
            l2.append(j)
print(len(l2))
s1=set(l2)
len(s1)
```

```python
df.isna().sum()
```

```python
df
```

```python
l3=[]
for i in s1:
    d={}
    d["country"]=i
    d["no_of_shows"]=len(df[(df["Country"].str.contains(i)==True) & (df["Category"]=="TV Show")])
    l3.append(d)
a=pd.DataFrame(l3)
a
    
```

```python
a.sort_values(by="no_of_shows",ascending=False).head(20)
```
