```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("hacker_news.csv")
df
```

* id: the unique identifier from Hacker News for the post
* title: the title of the post
* url: the URL that the posts links to, if the post has a URL
* num_points: the number of points the post acquired, calculated as the total number of upvotes minus the total number of downvotes
* num_comments: the number of comments on the post
* author: the username of the person who submitted the post
* created_at: the date and time of the post's submission

```python
df["id"].value_counts()
```

* Create three categories called ask_posts, show_posts, and other_posts

```python
ask_posts=df[df["title"].str.lower().str.startswith("ask hn")]
ask_posts
```

```python
show_posts=df[df["title"].str.lower().str.startswith("show hn")]
show_posts
```

```python
other_posts=df[(df["title"].str.lower().str.startswith("ask hn")==False) & (df["title"].str.lower().str.startswith("show hn")==False)]
other_posts
```

* Find the total number of comments in ask posts, and assign it to total_ask_comments

```python
ask_posts.dtypes
```

```python
total_ask_comments=ask_posts["num_comments"].sum()
total_ask_comments
```

* Compute the average number of comments on ask posts, and assign it to avg_ask_comments.

```python
ask_posts.isna().sum()
```

```python
avg_ask_comments=ask_posts["num_comments"].mean()
avg_ask_comments
```

* Find the total number of comments in show posts, and assign it to total_show_comments

```python
total_show_comments=show_posts["num_comments"].sum()
total_show_comments
```

* Compute the average number of comments on show posts and assign it to avg_show_comments

```python
avg_show_comments=show_posts["num_comments"].mean()
avg_show_comments
```

* Do show posts or ask posts receive more comments on average? 

* Finding the Amount of Ask Posts and Comments by Hour Created

```python
ask_posts["created_at"]=pd.to_datetime(ask_posts["created_at"])
```

```python
ask_posts
```

```python
ask_posts.groupby(ask_posts["created_at"].dt.hour)["num_comments"].sum() #comments by hour
```

```python
ask_posts.groupby(ask_posts["created_at"].dt.hour)["num_comments"].count() #counts by hour
```

 * Calculating the Average Number of Comments for Ask HN Posts by Hour

```python
ask_posts.groupby(ask_posts["created_at"].dt.hour)["num_comments"].mean() #avg comments by hour
```

* "Top 5 Hours for Ask Posts Comments"

```python
ask_posts.groupby(ask_posts["created_at"].dt.hour)["num_comments"].mean().sort_values(ascending=False).head(5)
```
