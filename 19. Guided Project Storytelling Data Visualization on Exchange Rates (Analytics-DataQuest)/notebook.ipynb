## 2.Data Cleaning

* We rename the [US dollar ] and Period\Unit: columns to something easier to type — US_dollar and Time.
* We change the Time column to a datetime data type.
* We sort the values by Time in ascending order.
* We reset the index (and drop the initial index).

```python
exchange_rates.rename(columns={"[US dollar ]":"US_dollar" , "Period\\Unit:" :  "Time"}, inplace=True)
```

```python
exchange_rates["Time"]=pd.to_datetime(exchange_rates["Time"])
```

```python
exchange_rates.sort_values(by="Time",inplace=True)
```

```python
exchange_rates.reset_index(drop=True,inplace=True)
```

* Isolate the Time and the US_dollar columns. Assign them to a different variable named euro_to_dollar.
* Run the Series.value_counts() method on the US_dollar column, and see if you notice anything wrong.
* Drop all the rows where the - character appears in the US_dollar column.
* Convert the US_dollar column to a float data type.

```python
euro_to_dollar=exchange_rates[["Time","US_dollar"]]
```

```python
euro_to_dollar["US_dollar"].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar["US_dollar"]!="-"]
```

```python
euro_to_dollar["US_dollar"]=euro_to_dollar["US_dollar"].astype(float)
```

## 3.Rolling Mean

```python
sns.lineplot(data=euro_to_dollar,x="Time",y="US_dollar")
```

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: : they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the rolling mean (also known as the moving average).

Refer this for detailed explanation about rolling mean
https://app.dataquest.io/c/96/m/529/guided-project%3A-storytelling-data-visualization-on-exchange-rates/3/rolling-mean

* Calculate the rolling means for the US_dollar column using a moving window of 30 days. Add the rolling means to a new column named rolling_mean

```python
euro_to_dollar["rolling_mean"] = euro_to_dollar["US_dollar"].rolling(window=30).mean()
```

```python
sns.lineplot(data=euro_to_dollar,x="Time",y="rolling_mean")
```

## Storytelling Data Visualization Examples

### Financial Crisis Example

```python
financial_crisis = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2006
                   ) & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy(
                   )[(euro_to_dollar.Time.dt.year >= 2007
                   ) & (euro_to_dollar.Time.dt.year <= 2008)]
```

```python
a=pd.to_datetime("2008-04-01")
b=pd.to_datetime("2008-09-01")
```

```python
### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')



### Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
sns.lineplot(x=financial_crisis['Time'],
        y=financial_crisis['rolling_mean'],
        linewidth=1, color='#A6D785')
plt.xticks(rotation=90)



### Highlighting the 2007-2008 period
sns.lineplot(x=financial_crisis_7_8['Time'],
        y=financial_crisis_7_8['rolling_mean'],
        linewidth=3, color='#e23d28')



### Highlihting the peak of the crisis
ax.axvspan(xmin=a,xmax=b,ymin=0.09,
           alpha=0.3, color='grey')
 

### Add some transparency to the grid
ax.grid(alpha=0.5)
```

### The Three US Presidencies Example

```python
bush_obama_trump = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
fig=plt.figure(figsize=(20,15))
ax1=plt.subplot(2,3,1)
ax1.set(title="BUSH 2001-2009")
sns.lineplot(data=bush,x="Time",y="rolling_mean",ax=ax1,color="red")
ax1.tick_params(axis='x', labelrotation = 90)
ax1.set(ylim=[0.8,1.6])

ax2=plt.subplot(2,3,2)
ax2.set(title="OBAMA 2009-2017")
sns.lineplot(data=obama,x="Time",y="rolling_mean",ax=ax2)
ax2.tick_params(axis='x', labelrotation = 90)
ax2.set(ylim=[0.8,1.6])

ax3=plt.subplot(2,3,3)
ax3.set(title="TRUMP 2017-2021")
sns.lineplot(data=trump,x="Time",y="rolling_mean",ax=ax3,color="green")
ax3.tick_params(axis='x', labelrotation = 90)
ax3.set(ylim=[0.8,1.6])


ax4=plt.subplot(2,3,(4,6))
sns.lineplot(data=bush,x="Time",y="rolling_mean",ax=ax4,color="red")
sns.lineplot(data=obama,x="Time",y="rolling_mean",ax=ax4)
sns.lineplot(data=trump,x="Time",y="rolling_mean",ax=ax4,color="green")

#setting signature
ax4.text(x=10957,y=0.8,s="prashu._.n",color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=20)

# ax4.get_xticks()
# ax4.get_yticks()
```
