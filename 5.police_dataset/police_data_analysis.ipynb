Q. 3) Question ( Groupby ) - Does gender affect who gets searched during a stop ?

```python
df
```

```python
df["stop_outcome"].value_counts()
```

```python
df["stop_duration"].value_counts()
```

```python
df.dtypes
```

```python
l=[]
for i,j in df.groupby(["driver_gender"],as_index=False)["search_conducted"]:
    d={}
    print(i,j.value_counts())
    d["Gender"]=i
    d["Percentage"]=j.value_counts()[True]/j.value_counts()[False]*100
    l.append(d)
print(l)
```

```python
a=pd.DataFrame(l)
ax=sns.barplot(data=a,x=a.Gender,y=a.Percentage)
for bars in ax.containers:
    ax.bar_label(bars)
```
