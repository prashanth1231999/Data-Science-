```python
#3 palindrome

s="aabbbaaa"
l=[]
for i in range(len(s)+1):
    for j in range(i+1):
        e=s[j:i]
        if e==e[::-1] and e!="":
            l.append(e)
c=0
c1=0
for i in l:
    if c1!=0:
        break
    for j in l:
        if c!=0:
            c1=c1+1
            break
        for k in l:
            if i+j+k==s:
                print(i,j,k)
                c=c+1
                break
                
```
