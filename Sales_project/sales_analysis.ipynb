```python
import os
l=os.listdir(".")
a=[]
for i in l:
    if i.endswith("csv"):
        a.append(pd.read_csv(i))
df=pd.concat(a,axis=0)
```

```python
df.replace({"Quantity Ordered":{"Quantity Ordered":np.nan}},inplace=True)
```

```python
df.replace({"Price Each":{"Price Each":np.nan}},inplace=True)
```

```python
df["Quantity Ordered"]=df["Quantity Ordered"].astype(float)
```

```python
df["Price Each"]=df["Price Each"].astype(float)
```

```python
df["total_price"]=df["Price Each"]*df["Quantity Ordered"]
```

```python
a=df.groupby(df.Month.dt.month)
```

```python
a.sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)
print(months)

plt.bar(months,a.sum()['total_price'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```
